```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
combined = train.drop('Survived', axis = 1).append(test)
combined.head()
```

```python
combined.Sex = combined.Sex.apply(lambda x: 1 if x == 'male' else 0)
combined.head()
```

```python
def Name_Title_Code(x):
    if x == 'Mr.':
        return 1
    if (x == 'Mrs.') or (x == 'Ms.') or (x == 'Lady.') or (x=='Mlle.') or (x == 'Mme.') or (x =='Dona.'):
        return 2
    if x =='Miss.':
        return 3
    if (x == 'Rev.') or (x == 'Don.') or (x=='Dr.') or (x=='Major.') or (x == 'Sir.') or x == 'Col.' or (x == 'Capt.') or (x == 'Jonkheer.'):
        return 4
    return 5    

combined['Name_Title'] = combined.Name.apply(lambda x: x.split(',')[1]).apply(lambda x: x.split()[0])
combined['Name_Title1'] = combined['Name_Title'].apply(Name_Title_Code)

del combined['Name']
combined.head()
```

```python
medMas = combined[combined['Name_Title1'] == 5].Age.median()
ag = combined[combined['Name_Title1'] != 5].Age.mean()
```

```python
combined.Age[combined['Name_Title1'] == 5] = combined.Age.fillna(medMas)
combined.Age = combined.Age.fillna(ag)
combined.Age = combined.Age.round(decimals = 1)
```

```python
combined.info()
```

```python
combined['Family_size'] = np.where((combined.SibSp + combined.Parch == 0), 'Single',
                                  np.where((combined.SibSp + combined.Parch < 3), 'Small', 'Big'))
del combined['SibSp']
del combined['Parch']
```

```python
combined['Name_Title'] = combined['Name_Title1']
del combined['Name_Title1']
```

```python
# Ticket

combined['Ticket_Lett'] = combined['Ticket'].apply(lambda x: str(x)[0])
k = pd.DataFrame([combined['Ticket_Lett'].groupby(combined.Pclass), combined.Pclass])
k.index = ['TicketLet', 'Class']
k
```

```python
combined['Ticket_Lett'] = combined['Ticket_Lett'].apply(lambda x: str(x))
combined['Ticket_Lett'] = np.where(combined['Ticket_Lett'].isin(['L', '9', '8', '7', 'A']), 'Low_Ticket', 
                                   np.where(combined['Fare'] > 60, 'Lux_Ticket', 'Medium_Ticket'))
```

```python
combined['Ticket_Lett'] = np.where(combined['Fare'] < 11, 'Low_Ticket', 'Medium_Ticket')
```

```python
combined[combined['Ticket_Lett'] == 'Medium_Ticket']
```

```python
combined.info()
```

```python
combined['Cabin_let'] = combined.Cabin.apply(lambda x: str(x)[0])
combined['Cabin_n1'] = combined.Cabin.apply(lambda x: str(x).split(' ')[-1][1:])
combined['Cabin_n1'].replace('an', np.NaN, inplace = True )
combined['Cabin_num'] = pd.qcut(combined['Cabin_n1'][:891],3)
combined = pd.concat((combined, pd.get_dummies(combined['Cabin_num'], prefix = 'Cabin_num')), axis = 1)
```

```python
cab = combined['Cabin_let'].astype('category').cat.categories
k3 = combined['Fare'].groupby(combined['Cabin_Let']).mean()
a = pd.Series([cab, k3], index = ['Cabin_Let Letter', 'Fare'])
#a['Fare'].sort_values(axis= 0, ascending=True)
```

```python
combined.Embarked.fillna('S', inplace = True)
```

```python
sns.countplot(x = combined.Embarked, data = combined )
```

```python

del combined['PassengerId']
```

```python
columns = ['Embarked', 'Ticket_Lett', 'Name_Title', 'Family_Size']
for column in columns:
    combined[column] = combined[column].apply(lambda x: str(x))
    good_cols = [column+'_'+i for i in combined[:891][column].unique() if i in combined[891:][column].unique()]
    combined = pd.concat((combined, pd.get_dummies(combined[column], prefix = column)[good_cols]), axis = 1)
del combined[column]
```

```python
targets = train.Survived
train = combined[:891]
test = combined[891:]
```
